```python
import os
from pathlib import Path
import shutil
import numpy as np
```

```python
# 读取原始数据文件名列表
carla_dir = 'data/carla_scene_02_unnorm'
target_dir = 'data/carla_scene_02'
file_list = os.listdir(carla_dir)
```

```python
# 创建生成数据文件目录
gen_dir = Path(target_dir)
gen_dir.mkdir(exist_ok=True)
source_folder = os.getcwd() + '/' + carla_dir
target_folder = os.getcwd() + '/' + target_dir
```

```python
for file in file_list:
    # shutil.copyfile(source_folder + file, target_folder + file)
    if file[-3:] == 'npz':
        raw_data = np.load(source_folder + '/' +file)['arr_0']
    else:
        raw_data = np.load(source_folder + '/' +file)
    max_record = np.zeros((4),dtype=float)
    min_record = np.ones((4),dtype=float)
    for raw in raw_data:
        for i in range(4):
            # xyzv的最大值记录
            max_record[i] = max(raw[i],max_record[i])
            min_record[i] = min(raw[i],min_record[i])
    renewdata = []
    for raw in raw_data:
        for i in range(4):
            raw[i] = (raw[i] - min_record[i]) / (max_record[i] - min_record[i])
        renewdata.append([raw[0], raw[1], raw[2],raw[3],raw[4],raw[5]])
    renewdata = np.asarray(renewdata)
    np.savez(target_folder +'/' + file[:-4], renewdata)
```
